```python
def arange(start, stop, step=1):
    N = int((stop - start) / step - 0.0000001) + 1
    print(f'N = {N}  from {(stop - start) / step}')
    return [start + i * step for i in range(N)]
```

```python
arange(-1, 1.00000001, 0.5)
```

```python
import numpy as np
import matplotlib.pyplot as plt
delta = 1.0
x = np.arange(-2.0, 2.0001, delta)
y = np.arange(-1.0, 1.0001, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=True, fontsize=10)
ax.set_title('Simplest default with labels')
```
